```python
import time
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import os
import requests
%config InlineBackend.figure_format = 'svg'
from pylab import rcParams

rcParams["figure.figsize"] = 8, 5

import numpy as np
import pandas as pd
```

```python
participants_path = '../../data/Участники anonimized.csv'
```

```python
participants = pd.read_csv('../../data/Участники anonimized.csv').rename(columns={'Место работы': 'workplace'})
```

```python
participants
```

```python
participants['workplace'] = participants['workplace'].fillna('-')
participants['workplace'] = participants['workplace'].str.replace('\'', '').str.replace('"', '').str.replace('»', '').str.replace('«', '')
participants['workplace']
```

```python
import requests

url = r'https://yandex.ru/maps/?from=tabbar&text='
header = 'analyticsId'


def parse(place):
    html = requests.get(url + place).text

    names_list = []
    coords_list = []
    pos = 0

    while html.find(header, pos) != -1:
        pos = html.find(header, pos) + len(header)
        name = html[html.find('title', pos) + 8:html.find('description', pos) - 3]
        coords = html[html.find('[', pos) + 1:html.find(']', pos)].split(',')
        coords = list(map(float, map(str.strip, coords)))
        name.replace('"', '')
        names_list.append(name)
        coords_list.append(coords)
    if len(names_list) == 0:
        return 'not_parsed', [0, 0]
    else:
        return names_list[0], coords_list[0]
```

```python
parsed = []

for i in range(3):
    name, coords = parse(participants.iat[i, 7])
    if name.lower() in parsed:
        continue
    participants.at[i, 'parsed_workplace'] = name
    participants.at[i, 'lon'] = coords[1]
    participants.at[i, 'lat'] = coords[0]
    time.sleep(10)
```
